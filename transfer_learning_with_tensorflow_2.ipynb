##**Processing the Image**

```python
import tensorflow as tf
import numpy as np
def process_image(image):
  img_process = tf.keras.preprocessing.image.img_to_array(image)
  img_process = np.expand_dims(img_process, axis= 0)
  final_image = tf.keras.applications.resnet50.preprocess_input(img_process)
  return final_image
```

```python
lion = process_image(lion)
elephant = process_image(elephant)
cat = process_image(cat)
dog = process_image(dog)
watch = process_image(watch)
bicycle = process_image(bicycle)

lion.shape, elephant.shape, cat.shape, dog.shape, watch.shape, bicycle.shape
```

```python
import tensorflow as tf

def image_prediction(image):
  model = tf.keras.applications.ResNet50(weights="imagenet")
  prediction = model.predict(image)
  decode_image = tf.keras.applications.resnet50.decode_predictions(prediction)

  result = print(f" Model Prediction : {decode_image[0][0][1]} \nConfedience Level : {decode_image[0][0][2] *100: 0.2f}%")

  return result
```

```python
image_prediction(lion)
```

##**Image Prediction with single function**

```python
import tensorflow as tf
import numpy as np

def raw_image_prediction(image):
  
  model = tf.keras.applications.ResNet50(weights="imagenet")
  image_process = tf.keras.preprocessing.image.load_img(image,target_size = (224,224))
  image_process = tf.keras.preprocessing.image.img_to_array(image_process)
  image_process = np.expand_dims(image_process, axis = 0)
  image_process = tf.keras.applications.resnet50.preprocess_input(image_process)
  prediction = model.predict(image_process)
  image_decode = tf.keras.applications.resnet50.decode_predictions(prediction)

  result = print(f"Model Prediction : {image_decode[0][0][1]} \nConfedience Level : {image_decode[0][0][2] * 100 :0.2f}%")

  return result
  





```

```python
raw_sample = "lion.jpg"
```

```python
raw_image_prediction(raw_sample)
```
